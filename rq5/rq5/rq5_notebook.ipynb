```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
compat_df = pd.read_csv(PATH_COMPAT)

# Calculate the distribution of values in the 'compatibilities' column
compatibility_distribution = compat_df['compatibilities'].value_counts()

# Sort the index (unique values) of the compatibility_distribution Series in ascending order
compatibility_distribution_sorted = compatibility_distribution.sort_index()

# Calculate the mean value of the 'compatibilities' column
mean_value = compat_df['compatibilities'].mean()

# Get the positions and labels for the x-axis
positions = range(len(compatibility_distribution_sorted))
labels = compatibility_distribution_sorted.index

# Plot the distribution
plt.figure(figsize=(8, 6))
compatibility_distribution_sorted.plot(kind='bar', color='skyblue')

# Find the closest position to the mean value to place the line correctly
closest_position = min(positions, key=lambda x: abs(labels[x] - mean_value))
plt.axvline(closest_position, color='red', linestyle='dashed', linewidth=1.5)
plt.text(closest_position, plt.ylim()[1] * 0.9, f'Mean: {mean_value:.2f}', color='red', fontsize=12, ha='right')

# Add grid
plt.grid(True, linestyle='--', alpha=0.7)

plt.title('Range size distribution over GAVs')
plt.xlabel('Number of versions in the compatible range')
plt.ylabel('Number of GAVs')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.savefig('rq5_range_size_distribution.eps', format='eps')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set seaborn style
# sns.set(style="whitegrid")

# Columns to include in the plot
columns_to_plot = ['Change rate', 'Change magnitude', 'Upgrades',
                   'Upgrade steps','Downgrades', 'Downgrade steps',
                   'Replacements', 'Replacement rate', 'Resolved']

# Calculate the number of rows and columns needed for subplots
num_columns = len(columns_to_plot)
num_cols = 2  # Ensure maximum 2 columns per row
num_rows = (num_columns + num_cols - 1) // num_cols  # Round up to the nearest integer

# Create subplots with the calculated number of rows and columns
fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, num_rows * 5))

# Flatten the axs array for easier indexing
axs = axs.flatten()

# Plot each distribution
for i, column in enumerate(columns_to_plot):
    # Plot the distribution on the corresponding subplot if there are more columns to plot
    sns.histplot(rows_with_replacements[column], kde=True, bins=25, color='skyblue', edgecolor='black', linewidth=1.5, ax=axs[i])

    # Calculate the mean value of the column
    mean_value = rows_with_replacements[column].mean()
    
    # Plot a vertical line for the mean value
    axs[i].axvline(mean_value, color='red', linestyle='dashed', linewidth=1.5)
    axs[i].text(mean_value, axs[i].get_ylim()[1] * 0.9, f'Mean: {mean_value:.2f}', color='red', fontsize=12)
    
    # Set title and labels
    title_name = "Resolved dependencies" if column == "Resolved" else column
    axs[i].set_title(f'Distribution of {title_name}', fontsize=16)
    axs[i].set_xlabel(title_name, fontsize=14)
    axs[i].set_ylabel('Number of projects', fontsize=14)
    axs[i].tick_params(axis='x', labelsize=12)
    axs[i].tick_params(axis='y', labelsize=12)
    axs[i].grid(True, linestyle='--', alpha=0.7)

# fig.delaxes(axs[-1]) # The indexing is zero-based here
# Remove any empty subplots

for j in range(i + 1, num_rows * num_cols):

    fig.delaxes(axs[j])

# Adjust layout and display the plots
plt.tight_layout()

plt.savefig('distributions.eps', format='eps')
plt.show()
```
