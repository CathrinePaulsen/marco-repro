```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# result_path = "./rq3_results.csv"
result_path = "./rq3_results_manual_eval.csv"
df = pd.read_csv(result_path).drop_duplicates(subset=["ga", "version"])
df
```

```python
# Count the number of rows where the 'repo' column has a non-null value
total = df.shape[0]
print(f"Total n datapoints: {total}")

num_rows_with_repo_value = df['repo'].count()
print(f"Number of rows with 'repo' value: {num_rows_with_repo_value} ({num_rows_with_repo_value/total*100}%)")

num_rows_with_tag_value = df['tag'].count()
print(f"Number of rows with 'tag' value: {num_rows_with_tag_value} ({num_rows_with_tag_value/total*100}%)")

num_true_exact_match = df[df['exact_match'] == True].shape[0]
print(f"Number of rows with 'exact_match' equals TRUE: {num_true_exact_match} ({num_true_exact_match/total*100}%)")

num_false_exact_match = df[df['exact_match'] == False].shape[0]
print(f"Number of rows with 'exact_match' equals False: {num_false_exact_match} ({num_false_exact_match/total*100}%)")

num_has_tests = df[df['has_tests'] == True].shape[0]
print(f"Number of rows with 'has_tests' equals True: {num_has_tests} ({num_has_tests/total*100}%)")

num_has_tests_jar = df[df['has_tests_jar'] == True].shape[0]
print(f"Number of rows with 'has_tests_jar' equals True: {num_has_tests_jar} ({num_has_tests_jar/total*100}%)")

print(f"Number of link failures: {total - num_rows_with_tag_value} ({(total - num_rows_with_tag_value)/total*100}%)")
```

```python
match_df = df[(df['tag'] == df['groundtruth_tag']) | (df['commit'] == df['groundtruth_tag'])]
matches = match_df.shape[0]
print(f"Number of correct matches: {matches} ({matches/total*100}%)")

match_df['repo'] = match_df['repo'].str.strip().str.lower()
match_df['groundtruth_repo'] = match_df['groundtruth_repo'].str.strip().str.lower()
diff_repos_df = match_df[match_df['repo'] != match_df['groundtruth_repo']]
print(f"Found {diff_repos_df.shape[0]} points of correct tag matches but different repos?")
# Manual check revealed that they different repo given ground truth redirects to the repo given by the linking algorithm,
# so matching on tag sha is sufficient.

# # Print the filtered DataFrame
# print("Rows where 'repo' equals 'groundtruth_repo' and 'groundtruth_tag' does not equal 'tag' or 'commit':")
# print(filtered_df)

wrong_match_df = df[(df['tag'] != df['groundtruth_tag']) & (df['commit'] != df['groundtruth_tag']) & (df['err'].isna())]
wrong_matches = wrong_match_df.shape[0]
print(f"Number of incorrect matches: {wrong_matches} ({wrong_matches/total*100}%)")

wrong_match_df.to_csv("./incorrect_matches.csv", index=False)

for index, row in wrong_match_df.iterrows():
    print(f"\n{row['ga']}:{row['version']}, gt_repo={row['groundtruth_repo']}, gt_tag={row['groundtruth_tag']}, repo={row['repo']}, tag={row['tag']}, commit={row['commit']}")
```
