```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
# Apply Seaborn style
sns.set(style="whitegrid")
```

```python
# Data
data = {
    "Category": ["static", "linked", "runnable", "dynamic"],
    "Total":        [371, 16, 3, 3],
    "Incompatible": [346, 0,  0, 1],
    "Compatible":   [25,  0,  0, 2]
}

# Adjust 'total' row to show the sum of 'Incompatible' and 'Compatible' counts
data["Incompatible"][0] = sum(data["Incompatible"])
data["Compatible"][0] = sum(data["Compatible"])

# Create the DataFrame
df = pd.DataFrame(data)

# Set the Category column as the index
df.set_index("Category", inplace=True)

# Plotting the stacked bar chart
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the 'Total' bars
ax.bar(df.index, df['Total'], label='Total', alpha=0.2)

# Overlay the 'Incompatible' and 'Compatible' bars
df[['Incompatible', 'Compatible']].plot(kind='bar', stacked=True, ax=ax)

plt.title('Incompatible vs Compatible vs Total')
plt.xlabel('Category')
plt.ylabel('Count')
# ax.set_yscale('log')
plt.legend(title='Type')
plt.xticks(rotation=45)
plt.show()
```
